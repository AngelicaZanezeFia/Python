## 3 - Encontrando e preenchendo valores em branco

```python
em_branco = np.nan # np.nan serve para identificar valores em branco

serie = Series(['linha 1', 'linha 2', em_branco, 'linha 4', 'linha 5', 'linha 6', em_branco, 'linha 8'])
serie
```

```python
serie.isnull() # .isnull() retornar uma valores booleanos (True ou Falso) se um item dentro de um objeto Pandas é nulo/vazio.
```

```python
np.random.seed(25)
df = DataFrame(np.random.randn(36).reshape(6,6))
df.loc[3:5, 0] = em_branco 
df.loc[1:4, 5] = em_branco
df
```

```python
df_preenchido = df.fillna(0) # .fillna() encontra os valores nulos do nosso objeto e preenche esses valores com o novo valor.
df_preenchido
```

Adicionar um dicionário para o método .fillna(). Usando um dicionário o método entende que cada chave do dicionário corresponde a uma coluna e o seu par o valor a ser substituido na respectiva coluna.

```python
dicio = {0: 0.1, 5: 1.25}
df_preenchido = df.fillna(dicio)
df_preenchido
```

## 4 - Contando e removendo valores em branco

```python
em_branco = np.nan
np.random.seed(25)
df = DataFrame(np.random.randn(36).reshape(6,6))
df.loc[3:5, 0] = em_branco
df.loc[1:4, 5] = em_branco
df
```

Para calcular a quantidade de valores nulos, por coluna, usamos o método .isnull() e o método .sum() a partir da matriz de valores Booleanos.

```python
df.isnull().sum() 
```

### Removendo valore sem branco


`.dropna()` - Para identificar e remover as linhas que contém qualquer valor nulo.

Para remover as colunas que apresentam valores nulos, basta passar `axis=1` como argumento para o método para que ele considere colunas ao invés de linhas.

```python
df_limpo = df.dropna(axis=1)
df_limpo
```

## 5 - Removendo dados duplicados

```python
dados = {
    'coluna 1': [1, 1, 2, 2, 3, 3, 3],
    'coluna 2': ['a', 'a', 'b', 'b', 'c', 'c', 'c'],
    'coluna 3': ['A', 'A', 'B', 'B', 'C', 'C', 'C'],
}

df = DataFrame(dados)
df
```

O método .duplicated() confere se cada linha é uma duplicata de uma linha anteriore cria uma Serie de valores True ou False indicando se a linha é uma linha duplicada ou nao no dataframe

```python
df.duplicated()
```

.drop_duplicates() - Para remover todas as linhas repetidas.

inplace - Para alterar o objeto ao invés de criar um novo.

```python
df.drop_duplicates()
```

```python
df
```

```python
df.drop_duplicates(inplace=True)
```

```python
df
```

Para remover linhas repetidas considerando apenas uma columna, passamos para o método .drop_duplicates() uma lista com o nome da coluna que ele deverá considerar.

```python
dados = {
    'coluna 1': [1, 1, 2, 2, 3, 3, 3],
    'coluna 2': ['a', 'a', 'b', 'b', 'c', 'c', 'c'],
    'coluna 3': ['D', 'A', 'B', 'B', 'C', 'C', 'C'],
}

df = DataFrame(dados)
df
```

```python
df.drop_duplicates(['coluna 3'])
```

## 6 - Concatenando dados

```python
df = pd.DataFrame(np.arange(36).reshape(6,6))
df
```

```python
df2 = pd.DataFrame(np.arange(15).reshape(5,3))
df2
```

O método concat() junta dados de duas fontes diferentes em uma única tabela. O comportamento padrão é juntar as fontes considerando os índices das colunas.

Passando o argumento axis=1 para o método, o Python entende que deve juntar essas fontes usando o índice das linhas de ambas fontes

```python
pd.concat([df, df2])
```

```python
pd.concat([df, df2], axis=1)
```

## 7 - Transformando dados

```python
df = pd.DataFrame(np.arange(36).reshape(6,6))
df
```

.drop() - Para remover linhas, indicando os índices para as linhas ou colunas que queremos remover.

NOTA: para remover colunas usar o argumento axis=1.

```python
df.drop([1,2], inplace=True)
```

```python
df
```

```python
df.drop([0,2], axis=1)
```

Acrescentando dados

```python
serie = Series(np.arange(6))
serie.name = "variavel_somada"
serie
```

```python
variavel_somada = DataFrame.join(df, serie) # .join() para unir dois datasets usando o índice das linhas ou usando uma coluna como chave.
variavel_somada
```

Para ordernar os valores das linhas de um DataFrame, de forma ascendente ou descentende, usamos o método .sort_values() passando o índice das colunas para ordenar os valores.

```python
df_ordenado = variavel_somada.sort_values(by=['variavel_somada'], ascending=[False])
df_ordenado
```

## 8 - Agrupamentos e agregações

```python
caminho = 'D:/Documents/1-CientistadeDados/Python/Python-Linkedin/dados/mtcars.csv'

carros = pd.read_csv(caminho, encoding = 'utf8')
carros.head()
```

Renomeando colunas

```python
carros.columns = ['nomes','mpg','cyl','disp', 'hpt', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
carros.head()
```

.groupby() - Para agrupar os dados de um DataFrame considerando uma coluna em particular, passamos o nome da coluna que queremos executar o agrupamento.

```python
coluna_agrupamento = carros['cyl']
grupos_carros = carros.groupby(coluna_agrupamento)
grupos_carros.mean()
```
