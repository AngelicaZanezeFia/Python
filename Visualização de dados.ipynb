## 4 - Legenda e anotações

```python
caminho = 'D:/Documents/1-CientistadeDados/Python/Python-Linkedin/dados/mtcars.csv'
carros = pd.read_csv(caminho)
carros.columns = ['nomes','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
mpg = carros['mpg']
```

Colocando títulos nos eixos

```python
x = range(1, 10)
y = [1, 2, 3, 4, 0.5, 4, 3, 2, 1]
plt.bar(x, y)

plt.xlabel('Nossa legenda no eixo X')
plt.ylabel('Nossa legenda no eixo Y')
plt.show()
```

Colocalndo legenda no gráfico

```python
import matplotlib.patches as mpatches

plt.plot(mpg)
plt.title('Consumo de gasolina')
plt.xlabel('nomes dos carros')
plt.ylabel('millhas/galão')
plt.xticks(mpg.index, carros.nomes, rotation=45) # para incerir o nome dos carros no eixo x e não números


legenda = mpatches.Patch(label='mpg') # mpg = milhas/galão
plt.legend(handles=[legenda])

plt.show()
```

Fazendo anotações no gráfico

```python
mpg.max() # consumo máximo max=agregador de máximo
```

```python
plt.plot(mpg)
plt.title('Consumo de gasolina')
plt.xlabel('nomes dos carros')
plt.ylabel('millhas/galão')
plt.xticks(mpg.index, carros.nomes, rotation=45)


legenda = mpatches.Patch(label='mpg')
plt.legend(handles=[legenda])

plt.annotate('Toyota Corolla', xy=(19,33.9), xytext = (21,33),
           arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
```

## 6 - Montando histogramas, box plots e scatter plots

```python
caminho = 'D:/Documents/1-CientistadeDados/Python/Python-Linkedin/dados/mtcars.csv'
carros = pd.read_csv(caminho)
carros.columns = ['nomes','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
carros.index = carros.nomes
mpg = carros['mpg']

plt.hist(mpg)
plt.plot()
plt.show()
```

```python
sns.distplot(mpg)
```

Scatterplots

```python
carros.plot(kind='scatter', x='hp', y='mpg', c=['darkgray'], s=150) # c = cor do gráfico, s = tamanho das bolinhas
```

```python
sns.regplot(x='hp', y='mpg', data=carros, scatter=True)
```

Gerando uma matriz de scatter plots

```python
sns.pairplot(carros)
```

```python
colunas = ['mpg', 'disp', 'hp', 'wt']
carros_df = pd.DataFrame((carros[colunas].values), columns = colunas)
carros_target = carros['am'].values

carros_df['group'] = pd.Series(carros_target, dtype='category')
```

```python
sns.pairplot(carros_df, hue='group', palette='hls')
```

Construindo boxplots

```python
carros.boxplot(column='mpg', by='am')
carros.boxplot(column='wt', by='am')
```

```python
sns.boxplot(x='am', y='mpg', data=carros, palette='hls')
```
